# Grouping

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
df.columns
```

```python
groups = df.groupby('year')
```

```python
for name, group in groups:
    print(f'Year: {name}')
    print(group.head())
```

```python
for name, group in df.groupby(['team', 'year']):
    print(f'Team: {name[0]}')
    print(f'Year: {name[1]}')
    print(group.head())
```

```python
for name, group in df.groupby(['year']):
    print(f'Year: {name}')
    print(group.head())
```
