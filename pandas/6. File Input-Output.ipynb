```python
data = pd.DataFrame({
    'Name': [
        'Param',
        'Dog',
        'Samosa'
    ],
    'Type': [
        'Person',
        'Animal',
        'Food'
    ]
}, index=['Pehla', 'Doosra', 'Teesra'])
data
```

```python
data.to_csv('ajeeb.csv', index=False)
# Name,Type
# Param,Person
# Dog,Animal
# Samosa,Food
```

```python
data.to_csv('ajeeb.csv')
# ,Name,Type
# Pehla,Param,Person
# Doosra,Dog,Animal
# Teesra,Samosa,Food
```

```python
data2 = pd.DataFrame({
    'obj': [
        'a',
        'b',
        'c'
    ],
    'count': [
        22,
        33,
        44
    ]
})
data2
```

```python
data.to_excel('ajeeb.xlsx')
```

```python
with pd.ExcelWriter('ajeeb2.xlsx') as writer:
    data.to_excel(writer, sheet_name='data')
    data2.to_excel(writer, index=False, sheet_name='data2')
```
